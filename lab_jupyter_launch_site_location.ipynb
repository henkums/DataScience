```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np
import pandas as pd

def haversine(lat1, lon1, lat2, lon2):
    # convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radius of Earth in km: 6371
    km = 6371 * c
    return km

launch_sites_df = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates().reset_index(drop=True)

distances = []

for i in range(len(launch_sites_df)):
    for j in range(i+1, len(launch_sites_df)):
        site1 = launch_sites_df.loc[i]
        site2 = launch_sites_df.loc[j]
        dist = haversine(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
        distances.append({
            'Site1': site1['Launch Site'],
            'Site2': site2['Launch Site'],
            'Distance_km': dist
        })

distances_df = pd.DataFrame(distances)
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
